```python
def sep_data(df, specific_question):
    '''
    '''
    data_columns = list(df.keys())
    cols = ['shift']
    count_cols = 1
    for data_column in data_columns:

        # print(data_column)
        if data_column in specific_question.split("*"):
            count_cols = count_cols + 1
            print(data_column, True, count_cols)
            cols.append(data_column)

        else:
            # print(data_column, False)
            continue
    return cols
```

```python
def tables_data(anova_specific_questions, physical_df, sleep_df):
    '''
    
    '''
    
    tables = []
    dfs = [physical_df, sleep_df]
    for df in range(len(dfs)):
        columns = []
        questions = []
        data_frame = dfs[df]
        # sep = []
        # print(type(data_frame))
        for anova_specific_question in range(len(anova_specific_questions)):
            specific_question = anova_specific_questions[anova_specific_question]
            # print(specific_question)
            
            cols = sep_data(data_frame, specific_question)
            # print("data : ", data)
            if len(cols) == 5:
                questions.append(specific_question)
                # questions.append(specific_question)
                columns.append(cols)
                # print("sep: ", sep)
        '''
        Start of integration
        '''
        tables.append(questions)
        tables.append(columns)
    return tables
```

```python
tables = tables_data(anova_specific_questions, physical_df, sleep_df)
tables
```

```python
def generate_table_two(questions, columns, df):
    
    for question in range(len(questions)):
        spec_ques = questions[question]
        cols = columns[question]
        print(question, spec_ques, cols)
        
        form_table = df.loc[0:, cols[0:]]
        print(form_table)
        
        activity = list(form_table.keys())[2]
        # print(activity)
        
        work_or_off = form_table.set_index("work")
        # print(work_or_off)
        
        at_work_df = work_or_off.loc["work"]
        # print(at_work_df)
                     
        off_work_df = work_or_off.loc["off"]
        # print(off_work_df)
        
        '''
        Start of integration
        '''
        
        r_objects=robjects.r
        r_objects.source("rAov.R")
        print("\n==============activity======================", activity)
        
        if activity == "rest":
            rr =r_objects.rest_model(at_work_df, off_work_df)
            # print(rr)
        elif activity == "step_ratio":
            step_ds = r_objects.step_model(at_work_df, off_work_df)
            # print(step_ds)
        elif activity == "vigorous_min":
            vigor = r_objects.vigorous_model(at_work_df, off_work_df)
            print(vigor)
```

```python
table_2_questions = tables[0]
table_2_columns = tables[1]
table_2_summary = generate_table_two(table_2_questions, table_2_columns, physical_df)
table_2_summary
```
